```python
x = [2, 3, 4, 6, 9] #temp
y = [1, 2, 5, 7, 9.6] #pressure
x1 = np.linspace(0, 10,100)
y1 = np.random.randint(0, 10, 100)
```

```python
plt.scatter(x1, y1, label = "random", c="r")
plt.plot(x, y,  label = "temp-press", c= "m")
plt.legend()
plt.title("Temp-pressure graph")
plt.xlabel("temp")
plt.ylabel("pressure")
plt.show()
```

```python
y1 = np.sin(x1)
z1 = np.cos(x1)
```

```python
plt.scatter(x1, y1, label = "Sin", c="b")
plt.plot(x1, z1,  label = "Cos", c= "m")
plt.legend()
plt.title("sin and cos graph")
plt.xlabel("redian")
plt.ylabel("sin and cos")
plt.show()
```
